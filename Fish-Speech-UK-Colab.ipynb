```python
# ▶️ Завантаження голосового зразка
import requests

url = "https://huggingface.co/datasets/AI-GPT/fish-speech-demos/resolve/main/uk_voice.wav"
with open("uk_voice.wav", "wb") as f:
    f.write(requests.get(url).content)
```

```python
# ▶️ Синтез українського мовлення
from inference import infer

text = "Привіт! Як у тебе справи сьогодні?"
speaker_wav = "uk_voice.wav"
lang = "uk"

output_path = "output_uk.wav"
infer(
    text=text,
    speaker_wav=speaker_wav,
    lang=lang,
    output_path=output_path
)

print(f"Готово! Файл: {output_path}")
```
